```python
df = sns.load_dataset('mpg')
df.head()
```

```python
print(df.describe())
```

```python
# 8
result = df.isnull().sum().sum()
print(result)
```

```python
# 9
df1 = df.copy()
avg = df['horsepower'].mean()
# print(avg)
df1['horsepower'] = df1['horsepower'].fillna(avg)
result = df1['horsepower'].median()
print(result)
```

```python
# 10
df1 = df.copy()
df1 = df1.dropna()
a50 = df1.head(int(len(df1) * 0.5))
print(a50['horsepower'].quantile(0.25))
```

```python
# 11
a3 = df[df['cylinders'] == 3]
a8 = df[df['cylinders'] == 8]
result = a3['mpg'].mean() - a8['mpg'].mean()
print(int(abs(result)))
```

```python
# 12  z-score 표준화 : z = (x-평균)/표준편차

a = df['horsepower'].std()
avg = df['horsepower'].mean()
z = (df['horsepower']-avg)/a
# print(z)

print(len(z[z>1.5]) + len(z[z<-1.5]))
```

```python
# 13 min-max-scaling = (x-min)/(max-min)
mms = (df['mpg'] - df['mpg'].min()) / (df['mpg'].max() - df['mpg'].min())
print(len(mms[mms > 0.8]))
```

```python
# 14
# 이상치 => ub = q3 + 1.5*IQR, lb = q1 - 1.5*IQR

q3, q1 = df['weight'].quantile(0.75), df['weight'].quantile(0.25)
iqr = q3 - q1
ub = q3 + 1.5 * iqr
lb = q1 - 1.5 * iqr

cond1 = df['weight'] > ub
cond2 = df['weight'] < lb

print(len(df[cond1 | cond2]))
```

```python
# 15
a80 = df[df['model_year'] > 80]
a = a80.sort_values(by = 'mpg', ascending = False)
print(a.head(1))
```

```python
# 16

u = df[df['origin'] == 'usa']['horsepower'].std()
e = df[df['origin'] == 'europe']['horsepower'].std()
print(round(abs(u-e), 1))
```
