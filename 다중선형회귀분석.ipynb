```python
# ☑️☑️문제☑️☑️
# 주어진 심혈관계 질환 데이터에서 몸무게(weight), 콜레스테롤 수치(cholesterol), 흡연 여부(smoke)를 활용하여 심혈관 질환 유무(cardio)를 예측하는 다중 선형 회귀모델을 만들려고 한다.
# 이때, cholesterol의 회귀 계수를 구하시오.
# (단, 회귀계수는 반올림하여 소수점 3자리까지 표시한다.)

# 출처 : https://www.youtube.com/watch?v=AlIJ_TPyXQ0
```

```python
# 데이터 로드
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/JEunJin/BigData_python/refs/heads/master/bigdata_csvfile/cardiovascular_heart_disease_data.csv')

# 데이터 확인
# print(df.info())
# print(df.head())

# 상수항 추가 및 모델 생성
import statsmodels.api as sm

X = sm.add_constant(df[['weight', 'cholesterol', 'smoke']])
# y = df['cardio']
model = sm.OLS(df['cardio'], X)

# 회귀분석
result = model.fit()

# 분석 결과 확인
print(result.summary())
```
