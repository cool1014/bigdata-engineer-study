```python
# 데이터 로드
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/JEunJin/BigData_python/refs/heads/master/bigdata_csvfile/cardiovascular_heart_disease_data.csv')

from scipy.stats import chi2_contingency
import pandas as pd

table = pd.crosstab(df['smoke'], df['cardio'])

chi2, p, dof, expected = chi2_contingency(table)

# 검정통계량
print("chi2 =", chi2)
# p-value,  p < 0.05 => 귀무가설 기각(연관성 있음)
print("p =", p)
```
